```python
import json

with open("data/metadata-v2.json", "r") as infile:
    metadata = json.load(infile)
```

```python
locations = {}
for i in metadata.keys():
    if metadata[i]['cell'] in locations:
        locations[metadata[i]['cell']].append([metadata[i]['lat'],metadata[i]['lng']])
    else:
        locations[metadata[i]['cell']] = [[metadata[i]['lat'],metadata[i]['lng']]]

locations
```

```python
centroids = {}
for i in locations.keys():
    lat = 0
    lng = 0
    for j in locations[i]:
        lat += j[0]
        lng += j[1]
    centroids[i] = [lat/len(locations[i]), lng/len(locations[i])]

centroids
```

```python
#calculate distance between two points into a dictionary
from haversine import haversine
import math

def distance(lat1, lng1, lat2, lng2):
    #return math.sqrt((lat1-lat2)**2+(lng1-lng2)**2)
    return haversine((lat1,lng1), (lat2,lng2))

distances = {}
for i in centroids.keys():
    distances[i] = {}
    for j in centroids.keys():
        if i != j:
            distances[i][j] = distance(centroids[i][0], centroids[i][1], centroids[j][0], centroids[j][1])
        else:
            distances[i][j] = 0

distances
```

```python
#export to json
import json

with open('data/distances.json', 'w') as outfile:
    json.dump(distances, outfile)
```

```python
#get maximum distance
max_distance = 0
for i in distances.keys():
    for j in distances[i].keys():
        if distances[i][j] > max_distance:
            max_distance = distances[i][j]
max_distance
```
